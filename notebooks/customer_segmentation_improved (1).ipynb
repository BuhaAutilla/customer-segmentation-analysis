# Customer Segmentation with K-Means

## Business Question
Can customers be grouped into meaningful segments to support targeted marketing strategies?

## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

## Load Dataset

```python
df = pd.read_csv('../data/dataset.csv')
df.head()
```

## Feature Selection

```python
features = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

## Feature Scaling

```python
scaler = StandardScaler()
X = scaler.fit_transform(features)
```

## Elbow Method (Choosing Optimal K)

```python
inertia = []
k_range = range(1,11)
for k in k_range:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X)
    inertia.append(model.inertia_)

plt.figure()
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

## Apply K-Means Clustering

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

## Cluster Visualization

```python
plt.figure()
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=df['Cluster'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Customer Segments')
plt.show()
```

## Automatic Insight Generation

```python
cluster_summary = df.groupby('Cluster')[['Age','Annual Income (k$)','Spending Score (1-100)']].mean()
cluster_summary
```

```python
for i,row in cluster_summary.iterrows():
    print(f'Cluster {i}: avg age {round(row["Age"],1)}, avg income {round(row["Annual Income (k$)"],1)}, avg spending {round(row["Spending Score (1-100)"],1)}')
```
