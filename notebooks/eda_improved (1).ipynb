# Customer Segmentation — Exploratory Data Analysis

## Business Question
Which customer characteristics influence spending behavior and what patterns exist in the data?

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv('../data/dataset.csv')
df.head()
```

## Dataset Overview

```python
df.info()
df.describe()
```

## Data Cleaning

```python
df.isnull().sum()
df = df.drop_duplicates()
```

## Age Distribution

```python
plt.figure()
df['Age'].hist()
plt.title('Age Distribution')
plt.show()
```

## Income Distribution

```python
plt.figure()
df['Annual Income (k$)'].hist()
plt.title('Income Distribution')
plt.show()
```

## Spending Score Distribution

```python
plt.figure()
df['Spending Score (1-100)'].hist()
plt.title('Spending Score Distribution')
plt.show()
```

## Income vs Spending Relationship

```python
plt.figure()
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Income vs Spending')
plt.show()
```

## Correlation Matrix

```python
corr = df[['Age','Annual Income (k$)','Spending Score (1-100)']].corr()
plt.figure()
plt.imshow(corr)
plt.title('Correlation Matrix')
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.show()
```

## Observations
- Inspect distributions
- Look for relationships between income and spending
- Identify potential customer groups
